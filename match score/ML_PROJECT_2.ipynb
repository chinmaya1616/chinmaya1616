```python
import pandas as pd

# Assuming you've already loaded the dataset into df
# For example:
# df = pd.read_csv('your_dataset.csv')

# 1. Total runs in the entire dataset
total_runs = ipl['runs'].sum()

# 2. Total runs by a specific batsman
batsman_name = 'BB McCullum'  # Replace with the batsman you're interested in
batsman_runs = ipl[ipl['batsman'] == batsman_name]['runs'].sum()

# 3. Total runs in a specific match
match_id = 1  # Replace with the match ID you're interested in
match_runs = ipl[ipl['mid'] == match_id]['runs'].sum()

# 4. Total runs in the last 5 overs (overs >= 15, assuming 20-over match)
last_5_overs_runs = ipl[ipl['overs'] >= 15]['runs'].sum()

# 5. Total runs by each batting team
team_runs = ipl.groupby('bat_team')['runs'].sum()

# Printing the results
print(f"Total runs in the dataset: {total_runs}")
print(f"Total runs scored by {batsman_name}: {batsman_runs}")
print(f"Total runs in match {match_id}: {match_runs}")
print(f"Total runs in the last 5 overs: {last_5_overs_runs}")
print("Total runs by each team:")
print(team_runs)
```
